```python
d = float(input("Insira a distância em KM a ser percorrida:"))
c = float(input("Insira quantos KM o seu carro anda por litro:"))
pg = float(input("Insira o preço da gasolina:"))
#processamento
lgn = d/c
gastoG = pg*lgn
res2 = gastoG*2
print("Você gastará para ir R$", gastoG, "Ou R$", res2, "para ir e voltar")
```

#Velocidade média do carro

image.png

```python
d = float(input("Insira a distância em KM do percurso percorrido:"))
hg = float(input("Insira as horas gastas para  terminar o percurso:"))
mg = float(input("Agora insira os minutos gastos para terminar o percurso:"))
ms = mg/60
tt = hg+ms
vm = d/tt
print("A velociadade média do carro é:", vm)
```
